```python
import pandas as pd
import random
```

```python
stars = pd.read_csv("assignment_ds.csv")
```

```python
stars
```

```python
def rand_index(limit):
    number = 0
    index_list = list(range(limit))
    random.shuffle(index_list)
    return index_list
def select_variables(stars,index_list,number_train):
    #0:6
    x_train = stars.iloc[index_list[:number_train],0:4]
    x_test = stars.iloc[index_list[number_train:],0:4]
    y_train = stars.iloc[index_list[:number_train], -1]
    y_test =  stars.iloc[index_list[number_train:], -1]
    return x_train, y_train, x_test, y_test
```

```python
index_list = rand_index(stars.shape[0])

x_train, y_train, x_test, y_test = select_variables(stars,index_list, 190)
x_test
```

```python
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(6, activation='softmax'))

#model.add(Dense(num_pixels, input_dim=num_pixels, kernel_initializer='normal',activation='relu'))
#model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
#model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
y = []
ls = []
for i in y_train:
    ls = [0 for i in range(6)]
    ls[i] = 1
    y.append(ls)
y = pd.DataFrame(y)
```

```python
model.fit(x_train, y, epochs=1000)
```

```python
# evaluamos el modelo
scores = model.evaluate(x_train, y)
 
    
y_t = []
ls = []
for i in y_test:
    ls = [0 for i in range(6)]
    ls[i] = 1
    y_t.append(ls)
y_t = pd.DataFrame(y_t)

print (model.predict(x_test).round())
print (y_t)
```
