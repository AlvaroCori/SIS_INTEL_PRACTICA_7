## Load of a Neuronal Network and saving of outputs 

In this part we load models that were saveds and we can compare with the datas of other csv in order to get differents results. 

```python
import numpy
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import np_utils
from keras.models import model_from_json
from clean_variables import *
from select_variables import *
seed = 7
numpy.random.seed(seed)
```

We load the datas only for compare the test part.
* We load the first dataset.
* Then the category columns.
* get the variables for x and the variable for y.
* separate in train and test parts.

```python
stars = pd.read_csv("assignment_ds.csv")
colors = ['whitish','red','pale yellow orange','white-yellow','blue','yellow-white','yellowish','yellowish white','orange-red','orange','white','blue white','blue-white']
spectral_classes = ['O', 'F', 'G', 'K', 'A', 'M', 'B']
types = stars.iloc[:,-1]
stars_inputs = get_inputs(stars,colors, spectral_classes)
#stars_inputs = get_best_inputs(stars)
index_list = rand_index(stars.shape[0])
x_train, y_train, x_test, y_test = select_variables(stars_inputs,types,index_list, 160,24)
```

We need the name of the files and we can load the models in a neuronal network with the weights of his neurones.

```python
name = "caso_tercer_experimento"
```

```python
json_file = open("./jsons/"+name+'.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

loaded_model.load_weights("./h_fives/"+name+".h5")

loaded_model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
```

In this part we only use the first csv to compare the tests rows.

```python
y_predicted_exp = loaded_model.predict(x_test).round()
```

```python
y=IntegerEncode2(y_train)
scores = loaded_model.evaluate(x_train, y)
y_t = []
ls = []
for i in y_test:
    ls = [0 for i in range(6)]
    ls[i] = 1
    y_t.append(ls)

y_t = pd.DataFrame(y_t)
y_predicted = loaded_model.predict(x_test).round()

stadistics(convert_at_type_array(y_predicted_exp),convert_at_type_array(y_t.to_numpy()))

print(convert_at_type_array(y_predicted_exp))
print(convert_at_type_array(y_t.to_numpy())) 
```

We load the new test model that is independent of the first csv but have similar columns for the variables of x.

```python
stars_comp = pd.read_csv("test_ds.csv")
```

```python
colors = ['whitish','red','pale yellow orange','white-yellow','blue','yellow-white','yellowish','yellowish white','orange-red','orange','white','blue white','blue-white']
spectral_classes = ['O', 'F', 'G', 'K', 'A', 'M', 'B']
stars_comp = get_inputs(stars_comp,colors,spectral_classes)
stars_comp
```

```python
y_predicted_comp = loaded_model.predict(stars_comp).round()
y_predicted_comp = convert_at_type_array(y_predicted_comp)

print(y_predicted_comp)
```

Then we got the classification, we can create a new DataFrame that contain a Id column and the Type that the neuronal network have choose how a output.

```python
indexs = range(1,len(y_predicted_comp)+1)
```

```python
solution = pd.DataFrame()
solution["Id"] = indexs
solution["Type"] = y_predicted_comp
solution.set_index("Id")
```

The solution is saved in a csv file.

```python
#submission.csv
solution.set_index("Id").to_csv("submission.csv",encoding="utf-8")
```
