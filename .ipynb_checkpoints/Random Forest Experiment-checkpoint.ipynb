# Introduction
Random forest is a Supervised Machine Learning Algorithm that is used widely in Classification and Regression problems. It builds decision trees on different samples and takes their majority vote for classification and average in case of regression.

One of the most important features of the Random Forest Algorithm is that it can handle the data set containing continuous variables as in the case of regression and categorical variables as in the case of classification. It performs better results for classification problems.

# How random forest work

Random Forest is a type of Assembly in Machine Learning where we will combine different trees -we will see how and with what characteristics- and the output of each will be counted as "one vote" and the most voted option will be the response of the forest.

The steps are the next:

* We select certain columns from the totals where we create a tree.

* We create several trees by varying the columns selected in the previous step and also the number of samples for trees, this is known as "bootstrap sample".

* All trees are classified and we get their exits.

* We selected the results with the most votes for our result.


```python
import numpy
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import random
import matplotlib.pyplot as plt
import csv 
import seaborn as sns
seed = 7
numpy.random.seed(seed)
```

```python
train = pd.read_csv("assignment_ds.csv")
colors = ['whitish','red','pale yellow orange','white-yellow','blue','yellow-white','yellowish','yellowish white','orange-red','orange','white','blue white','blue-white']
spectral_classes = ['O', 'F', 'G', 'K', 'A', 'M', 'B']
```

This function allow convert a category text to a number for the reason that we can insert in the neuronal network 2 new columns.

```python
def ListCagetorietoNumber(colum,list_):
    pos=0
    _number=[]
    for x in colum:
        pos=0
        for y in list_:
            if(x.lower() == y.lower()):
                _number.append(pos)
            pos+=1
    return _number
train['Number_Color']=ListCagetorietoNumber(train.Color,colors)
train['Numbr_Spectral_Class']=ListCagetorietoNumber(train.Spectral_Class,spectral_classes)
train
```

We drop the columns of category because they has represented for other 2 columns.

```python
train=train.drop(['Color','Spectral_Class'],axis=1)
starts = train.pop('Type')
train.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, starts, test_size=0.25)
```

```python
x_train.head()
```

We call a class of a library for random forest and we realize in differents function the fit and the predict.

```python
model = DecisionTreeClassifier(random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
comp = pd.DataFrame({'Real':y_test,'prediction':y_pred})
```

We get the stadistic in base of the test cases.

```python
def stadistics(y_real, y_predicted):
    counter = 0
    total = len(y_real)
    for i in range(total):
        if (y_real[i] == y_predicted[i]):
            counter += 1
    acc= accuracy_score(y_real,y_predicted)
    print(f"Cantidad total de pruebas: {total}")
    print(f"Cantidad total de aciertos: {counter}")
    print(f"Porcentaje de aciertos: {counter/total*100}%")
    print(f"Accuracy: {acc}")
```

```python
stadistics(y_pred,y_test.array)
```

* Dump the trained decision tree classifier with Pickle
* Open the file to save as pkl file
* Close the pickle instances

```python
import pickle

decision_tree_pkl_filename = 'decision_tree_classifier_20170212.pkl'
 
decision_tree_model_pkl = open(decision_tree_pkl_filename, 'wb')
pickle.dump(model, decision_tree_model_pkl)
 
decision_tree_model_pkl.close()
```

Loading the saved decision tree model pickle

```python
decision_tree_model_pkl = open(decision_tree_pkl_filename, 'rb')
decision_tree_model = pickle.load(decision_tree_model_pkl)
print("Loaded Decision tree model :: ", decision_tree_model)
```

```python
stars_comp = pd.read_csv("test_ds.csv")

stars_comp['Number_Color']=ListCagetorietoNumber(stars_comp.Color,colors)
stars_comp['Numbr_Spectral_Class']=ListCagetorietoNumber(stars_comp.Spectral_Class,spectral_classes)
stars_comp=stars_comp.drop(['Color','Spectral_Class'],axis=1)

stars_comp.head()
```

We do the predict with the other test csv file.

```python
y_predicted_comp=decision_tree_model.predict(stars_comp)
y_predicted_comp
```

```python
indexs = range(1,len(y_predicted_comp)+1)
```

```python
solution = pd.DataFrame()
solution["Id"] = indexs
solution["Type"] = y_predicted_comp
solution.set_index("Id")
```

#### Limitations of Decision Trees

Decision trees have a tendency to over-adjust (overfit). This means that they tend to learn training data very well but their generalization is not so good. One way to improve the generalization of decision trees is to use regularization. In order to further improve the generalization capacity of decision trees, we will need to combine several trees.
